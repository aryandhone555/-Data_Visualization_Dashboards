```python
# prompt: Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about
# the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we
# can find any patterns in the data.

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load the Titanic dataset
df = sns.load_dataset("titanic")
```

```python
# prompt:  Plot a box plot for distribution of
# age with respect to each gender along with the information about whether they survived or
# not. (Column names : 'sex' and 'age')

# Create a box plot for age distribution with respect to gender and survival
sns.boxplot(x="sex", y="age", hue="survived", data=df)

# Add a title and labels to the plot
plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

# Display the plot
plt.show()
```
